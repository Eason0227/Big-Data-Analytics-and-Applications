```python
import re
import os 
import pandas as pd
import numpy as np
import time
```

### 讀取good的所有檔案，並合併成Data frame

```python
#need to put all csv in the same folder
path ='C:/Users/User/Desktop/good1'   #the csv file of location
files = os.listdir(path)               
files_csv = list(filter(lambda x: x[-4:]=='.csv' , files))

data_list = []
num_filter = re.compile(r'\d+')

for file in  files_csv:
    tmp = pd.read_csv(path + "/"+file, header = None)
    tmp['rank'] = num_filter.findall(file)[0]
    data_list.append(tmp)
    
all_data = pd.concat(data_list)

#tran the data to our need form
def trantodataframe(all_data):
    category_name = all_data[1]
    category_id = all_data[3]
    item_id = all_data[5]
    price = all_data[7]
    item_name = all_data[9]
    seller_nickname = all_data[11]
    seller_id = all_data[13]
    data = {'category_name':category_name,
       'category_id':category_id,
       'item_id':item_id,
       "price":price,
       "item_name":item_name,
       "seller_nickname ":seller_nickname ,
       "seller_id":seller_id}
    goodsdata = pd.DataFrame(data)
    return goodsdata
```

```python
data = trantodataframe(all_data)
data = data.dropna(subset=['item_name'], how='all') #移除空值
data = data.reset_index() #重新編碼
data["item_name"] = data["item_name"].astype(str) #全部轉成str
data['type'] = 'c'
```

### step 1. 以item name區分產品

```python
#輸入關鍵字
keyword = input("Input any text:") 
```

```python
star = time.time()
text_group_index = []
not_text_group_index = []

for j in range(len(data)):
    if (keyword in data["item_name"][j] or keyword in data["category_name"]):
        text_group_index.append(j)
    else:
        not_text_group_index.append(j)

end = time.time()
tot = end-star
print("time:",tot)
```

```python
#把每個產品標記他是text_group或不是text_group
data['type'][text_group_index] = 'text_group'
data['type'][not_text_group_index] = 'not_text_group'

text_group = data[ data['type']=='text_group']
```

### step3 判斷item name詞性組合  
### step4 判斷斷詞結果與關鍵字是否完全符合

N+N(key word)+N 配件  
N(key word)+N 配件  
N+space 主產品  
N+FW 主產品

```python
#pip install ckiptagger
#pip install tensorflow
#pip install gdown
from ckiptagger import data_utils
#data_utils.download_data_gdown("./")

from ckiptagger import WS, POS, NER
ws = WS("./data")
pos = POS("./data")
ner = NER("./data")
```

```python
text_group2 = text_group[text_group['product_type']=='Accessories']
text_group2
```

```python
main_product_index2 = []
Accessories_index2 = []
not_defined = []

for i in range(len(text_group2)):
    item_name  = text_group2.iloc[i,5]
    ws_results = ws([item_name])
    pos_results = pos(ws_results)

    for j in range(len(ws_results[0])):
        if  keyword in ws_results[0][j] : #關鍵字有出現在item name的斷詞裡
            index = j #關鍵字的位置

            if ws_results[0][index] == keyword: #step 3.關鍵字與item name的斷詞結果完全符合

                if index == len(ws_results[0])-1: #關鍵字出現在最後一個
                    main_product_index2.append(i)

                elif (pos_results[0][index+1] =='Na' or (pos_results[0][index-1] =='Na'and pos_results[0][index+1] =='Na')
                        or pos_results[0][index+1] !='WHITESPACE' or pos_results[0][index+1] !='FW') : #Step4. 關鍵字後面的字為名詞
                    Accessories_index2.append(i)
                else:
                    main_product_index2.append(i)
                    
            else:
                Accessories_index2.append(i) #step 3.關鍵字與item name的斷詞結果不完全符合

            break #可能一個item  name會出現多個關鍵字，只採用第一個關鍵字
        else :
            not_defined.append(i)
    #print(i)
```

```python
text_group2.iloc[main_product_index2,9] = 'Main_product'
text_group2.iloc[Accessories_index2,9] = 'Accessories'
```
